```python
# Import packages

import pandas as pd
from pandasql import sqldf
from pandasql import load_births

births = load_births()

print(sqldf("SELECT * FROM births where births > 200000 limit 5;", locals()))
```

```python
q = """ select date(date) as DOB, 
sum(births) as "Total Births" 
from births 
group by date 
limit 10; """

print(sqldf(q, locals()))
```

```python
print(sqldf(q, globals()))  # seem to have same effect as local()
```

```python
# creating a function
def pysqldf(q):
    return sqldf(q, globals())
```

```python
print(pysqldf(q))
```
